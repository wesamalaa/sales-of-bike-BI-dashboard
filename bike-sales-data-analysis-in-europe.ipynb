**In this Notebook, We are going to analyse the Bike sales data in European Countries to draw the future Inisghts of the Bike Sales Data.**

```python
#Importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Briefing****

1.Date: The Date on which Sales has Done.(Day,Month,Year).

2.Customer_Age: Represents the age of the Customer who do purchases.

3.Age_Group : The Age Group of the customer belongs to.

4.Customer_gender: Gender of the Customer.

5.Country  : The name of the country which customer belongs to.

6.State : The name of the state which customer belongs to.

7.Product_category : The category of the Sold product.

8.Sub_category : The sub_category of the product.

9.Order_qunatity: Quantity of the Orders done .

10.Unit_Cost: The unit cost of the each product.

11.Unit_Price : The unit Price of each product.

12.Profit : Net profit of the each Product.

13. cost : The total cost of the Order Qunatity of each Customer

14.Revenue: The total revenue of the each customer.
 


```python
#reading the CSV File
df =pd.read_csv("/kaggle/input/bike-sales-in-europe/Sales.csv")
df.head()
```

```python
# Identify the Data of Columns
df.info()
```

**Data Cleaning**

```python
df.describe()
```

```python
#Identify the Null Values and missing Values
df.isnull().sum()
```

```python
#Convert the datetime from object to datetime
import datetime as dt

df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
# check for missing values

df.isna().values.any()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.duplicated().values.any()
```

```python
#how many country,state,product,product_Category ? 

df[['Country','State','Product_Category','Product']].nunique()
```

```python
#how many unique countries names 

df['Country'].unique()
```

```python
#how many product category with names 

df['Product_Category'].unique()
```

```python
df['Customer_Gender'].replace('M','Male', inplace=True)
df['Customer_Gender'].replace('F','Female', inplace=True)
df.head(2)
```

```python
# Deleting the columns with wrong Calculations

df = df.drop(['Profit', 'Cost', 'Revenue'], axis =1)


```

```python
df.head()
```

```python
#Recalculating  the columns, and appends it to the data

df['Total_Cost'] = df['Order_Quantity'] * df['Unit_Cost']
df['Revenue'] = df['Order_Quantity'] * df['Unit_Price']
df['Profit'] = df['Revenue'] - df['Total_Cost']
df.head(2)
```

```python
age_group_df=df['Age_Group'].value_counts()
age_group_df
```

 Analysing the Bike Sales Data:

1. Age Distribution of Customers
2. Most Profitable Year
3. Most Profitable Month
4. Gender Having most ordered
5. Country/State Having highest revenue
6. category/Sub-Category generates most profit.

```python
## Analysing the distribution Using the Bar chart

age_group_df=df['Age_Group'].value_counts()
data1 = list(age_group_df.index)

# Set the width and height of the figure
plt.figure(figsize=(8,4))

# Add title
plt.title("Distribution of bike Consumers by Age-Group ",fontsize=18)

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=data1, y=age_group_df)

for i in range(len(age_group_df)):
    plt.text(i,age_group_df[i],age_group_df[i],ha='center',va='bottom')
    

# Add label for vertical axis
plt.ylabel("Total Number of Customers",fontsize= 14)
plt.show()
```

```python
sns.set_style("dark")
  
# plotting data on chart
plt.pie(age_group_df, labels=data1, autopct='%.0f%%')
plt.title('Distribution of Bike consumers by Age-Group')
  
# displaying chart
plt.show()
```

**Most Profitable Year**

```python
grouped_year_df = df.groupby('Year').sum()['Profit'].sort_values(ascending=False)
```

```python
grouped_year_df.plot(kind='barh', color='purple', alpha=0.5, title='Highest Profit Year-wise')
plt.show()
```

**Most Profitable Month**

```python
grouped_month_df = df.groupby('Month').sum()['Profit'].sort_values(ascending=False).head(5)
```

```python
grouped_month_df.plot(kind='bar', color='purple', alpha=0.5, title='Highest Profit Year-wise', rot=0, figsize=(8,4))
# plt.bar(grouped_month_data.index, grouped_month_data)
for i in range(len(grouped_month_df)):
    plt.text(i,grouped_month_df[i],grouped_month_df[i],ha='center',va='bottom')
plt.title('Top 5 Months by Profit')    
plt.show()
```

**Gender Having most ordered**


```python
Grouped_gender=df.groupby('Customer_Gender').sum()['Order_Quantity'].sort_values(ascending=False)
gender = list(Grouped_gender.index)
count = list(Grouped_gender.values)
```

```python
plt.pie(count, labels=gender,  autopct='%.0f%%')
plt.title('Distribution of Order Quantity by Gender')
  
# displaying chart
plt.show()
```

**Country/State Having highest revenue**


```python
grouped_country = df.groupby('Country').sum()['Revenue'].sort_values(ascending=False)
```

```python
grouped_country.plot(kind='bar', color='purple', alpha=0.5, rot=0, figsize=(8,4))

for i in range(len(grouped_country)):
    plt.text(i,grouped_country[i],grouped_country[i],ha='center',va='bottom')
plt.title('Highest Revenue Country-wise',fontsize=14)    
plt.show()
```

```python
grouped_country.plot(kind='pie', autopct='%.0f%%', title='Distribution of Revenue by Country')
plt.show()
```

```python
grouped_state = df.groupby('State').sum()['Revenue'].sort_values(ascending=False).head(5)
```

```python
grouped_state.plot(kind='pie', autopct = '%.0f%%', title='Distribution of Revenue by State')
plt.show()
```

```python
grouped_state.plot(kind='bar', color='purple', alpha=0.5, rot=0, figsize=(8,6))

for i in range(len(grouped_state)):
    plt.text(i,grouped_state[i],grouped_state[i],ha='center',va='bottom')
plt.title('Highest Revenue State-wise',fontsize=14)    
plt.show()
```

**category/Sub-Category generates most profit.**

```python
Grouped_product = df.groupby('Product_Category').sum()['Profit'].sort_values(ascending=False)
```

```python
Grouped_product.plot(kind='bar', rot=0, title='Profit by Product Category')
for i in range(len(Grouped_product)):
    plt.text(i, Grouped_product[i], Grouped_product[i], ha='center', va = 'bottom')
plt.show()
```

```python
grouped_subcategory = df.groupby('Sub_Category').sum()['Profit'].sort_values(ascending=False).head(5)
```

```python
grouped_subcategory.plot(kind='bar', rot=0, figsize=(10,4
                                                    ))
plt.title('Profit by Sub-Category', fontsize=14)
for i in range(len(grouped_subcategory)):
    plt.text(i, grouped_subcategory[i], grouped_subcategory[i], ha='center', va='bottom')
plt.show()
```
